```python
url = "https://www.imdb.com/chart/top/"
response = requests.get(url)
```

```python
# utworzenie DOM (Document Object Model)
import bs4

html_data = bs4.BeautifulSoup(response.text, features="lxml")
```

```python
tbody = html_data.find("tbody", class_="lister-list")
type(tbody)
```

```python
rows = tbody.find_all("tr")
```

```python
len(rows)
```

```python
rows[1]
```

```python
movies = []
for row in rows:
    rank = row.find("span")["data-value"].strip()
    print("rank:", rank)
    
    title = row.find("td", class_="titleColumn").find("a").text.strip()
    print("title:", title)
    
    link = row.find("td", class_="titleColumn").find("a")["href"].strip()
    print("link:", link)
    
    year = row.find("td", class_="titleColumn").find("span").text[1:-1].strip()
    print("year:", year)
    
    rating = row.find("td", class_="ratingColumn").text.strip()
    print("rating:", rating)
    
    print("="*40)
    
    movie = {
        "rank" : rank, "title" : title, "link" : link, "rating" : rating
    }
    movies.append(movie)
```

```python
df = pd.DataFrame(movies)
df.set_index("rank", inplace=True)
df["link"] = "https://imdb.com" + df.link
df
```
